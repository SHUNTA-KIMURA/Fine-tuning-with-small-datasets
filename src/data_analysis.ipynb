```python
import matplotlib.pyplot as plt
import numpy as np

# All values aggregated across users
fine_all = [0.6, 0.6, 0.6, 0.6, 0.2, 0.2, 0.4]
ref_all  = [0.4, 0.6, 0.6, 0.6, 0.4, 0.2, 0.4]

# Mean and std
means = [np.mean(fine_all), np.mean(ref_all)]
stds  = [np.std(fine_all),  np.std(ref_all)]

labels = ["Fine-tuned (DPO)", "Reference Model"]
x = np.arange(len(labels))

plt.figure(figsize=(5,4))
plt.bar(x, means, yerr=stds, capsize=6)
plt.xticks(x, labels)
plt.ylabel("Accuracy")
#plt.title("DPO vs Reference (Aggregated over Users)")
plt.tight_layout()
plt.show()
```
